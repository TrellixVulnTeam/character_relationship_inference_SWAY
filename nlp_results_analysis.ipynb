```python
import matplotlib.pyplot as plt
import pandas as pd
import operator
```

## FRIEND

```python
# friend_list = friend.read().split('\n')
friend = open('/content/top_friend.txt')
foe = open('/content/top_foe.txt')

temp = (friend.read())
# temp = (foe.read())
friend_list = temp.split('\n\n')
print(friend_list)
relations = {}
bla = []
for relation in friend_list:
  # print(relation)
  if 'Harry' in relation or 'Ron' in relation or 'Dumbledore' in relation or 'Snape' in relation or 'Malfoy' in relation:
    bla.append(relation.split('\n')[0])
# print(bla)

friend.close()
```

```python
parts = []
score = ''
relations_count = {}
relations_avgScore = {}
relations_Score = {}
for each in bla:  
  parts = each.split()
  score = parts.pop()
  parts.pop(0)
  # print(parts)
  if str(parts).strip("'") in relations_count:
    relations_count[str(parts).strip("'")] +=1
    relations_avgScore[str(parts).strip("'")] += float(score[6:])
    relations_Score[str(parts).strip("'")] = float(score[6:])
  elif str(parts[::-1]).strip("'") in relations_count:
    relations_count[str(parts[::-1]).strip("'")] +=1
    relations_avgScore[str(parts[::-1]).strip("'")] += float(score[6:])
    relations_Score[str(parts).strip("'")] = float(score[6:])
  else:
    relations_count[str(parts).strip("'")] = 1
    relations_avgScore[str(parts).strip("'")] = float(score[6:])
    # relations_Score[str(parts).strip("'")] = float(score[6:])
  # print('original ', str(parts).strip("'"))
  # print('reversed ', str(parts[::-1]).strip("'"))

  

print(relations_count)
print(relations_avgScore)
print(relations_Score)

```

```python
sorted_relation_c = dict( sorted(relations_count.items(), key=operator.itemgetter(1),reverse=True))
keys_c = sorted_relation_c.keys()
values_c = sorted_relation_c.values()
 
name = list(keys_c)
price = list(values_c)
 
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(name, price, color='#d3a625')
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
ax.xaxis.label.set_color('#740001')
ax.yaxis.label.set_color('#740001')

# Remove x, y Ticks

# ax.xaxis.set_ticks_position('none')
# ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5, color = '#740001')
ax.yaxis.set_tick_params(pad = 10, color = '#740001')
ax.tick_params(axis='x', colors='#740001')
ax.tick_params(axis='y', colors='#740001')


# Add x, y gridlines
ax.grid(b = True, color ='#740001',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='#740001')
 
# Add Plot Title
ax.set_title('Friend: Frequency of Relations Found',
             loc ='left', color = '#740001', fontsize = 20 )

 
# Show Plot
plt.savefig('Friend_freq.png')
plt.show()
```
